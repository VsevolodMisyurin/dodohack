### Создание фичи NewClient1 для всех ID, которые хоть раз имели 1

```python
orders['NewClient1'] = orders.groupby('ClientUUId')['NewClient'].transform('max')
orders['NewClient1'] = (orders['NewClient1'] > 0).astype(int)
```

```python
orders.NewClient1.value_counts()
```

### Создание данных медианы, минимума и максимума на основании OrderTotalPrice для каждого ID клиента

```python
orders['OrderMedianPrice'] = orders.groupby('ClientUUId')['OrderTotalPrice'].transform('median')
```

```python
orders['OrderMinPrice'] = orders.groupby('ClientUUId')['OrderTotalPrice'].transform('min')
orders['OrderMaxPrice'] = orders.groupby('ClientUUId')['OrderTotalPrice'].transform('max')
```

### Запишем медиану разницы между стоимостью в меню и фактической после всех скидок для всех товаров каждого клиента
Запишем её в процентах

```python
orders['DiscountPercent'] = (orders['MenuPrice'] - orders['ProductTotalPrice']) / orders['MenuPrice'] * 100
orders['MedianDiscountPercent'] = orders.groupby('ClientUUId')['DiscountPercent'].transform('median')
```

```python
orders = orders.drop(['DiscountPercent'], axis=1)
```

### Занесём данные о количестве уникальных ID продуктов, которые когда-либо были заказаны клиентом

```python
unique_order_counts = orders.groupby('ClientUUId')['ProductUUId'].nunique().reset_index(name='ProductUUIdCount')
orders = orders.merge(unique_order_counts, on='ClientUUId', how='left')
```

### Занесём данные о медиане числа уникальных ID продуктов в каждом индивидуальном заказе клиента

```python
# Подсчитываем количество уникальных ProductUUId для каждого OrderUUId и ClientUUId
unique_product_counts = orders.groupby(['ClientUUId', 'OrderUUId'])['ProductUUId'].nunique().reset_index(name='UniqueProductCount')
# Вычисляем медиану количества уникальных ProductUUId для каждого ClientUUId
median_product_counts = unique_product_counts.groupby('ClientUUId')['UniqueProductCount'].median().reset_index(name='ProductUUIdMedian')
orders = orders.merge(median_product_counts, on='ClientUUId', how='left')
```

### Сохраним информацию о самом частовстречающемся addressId для пользователя

```python
# Подсчитываем самое частое значение addressId для каждого ClientUUId, игнорируя NaN
platform_mode = orders.groupby('ClientUUId')['addressId'].agg(lambda x: x.dropna().mode()[0] if not x.dropna().empty else None).reset_index(name='addressIdMain')
orders = orders.merge(platform_mode, on='ClientUUId', how='left')
```

### Сохраним информацию о самом частовстречающемся deliverySectorId для пользователя

```python
# Подсчитываем самое частое значение deliverySectorId для каждого ClientUUId, игнорируя NaN
platform_mode = orders.groupby('ClientUUId')['deliverySectorId'].agg(lambda x: x.dropna().mode()[0] if not x.dropna().empty else None).reset_index(name='deliverySectorIdMain')
orders = orders.merge(platform_mode, on='ClientUUId', how='left')
```

### Запишем также максимальный номер заказа

```python
max_order_number = orders.groupby('ClientUUId')['ClientOrderNumber'].max().reset_index(name='ClientOrderMaxNumber')
orders = orders.merge(max_order_number, on='ClientUUId', how='left')
```

### Сохраним информацию о самой частовстречающейся категории товара для пользователя

```python
сategoryid_mode = orders.groupby('ClientUUId')['CategoryId'].agg(lambda x: x.mode()[0]).reset_index(name='CategoryIdMain')
orders = orders.merge(сategoryid_mode, on='ClientUUId', how='left')
```

### Сохраним самый распространённый тип ордера пользователя

```python
ordertype_mode = orders.groupby('ClientUUId')['OrderType'].agg(lambda x: x.mode()[0]).reset_index(name='OrderTypeMain')
orders = orders.merge(ordertype_mode, on='ClientUUId', how='left')
```

### Сохраним данные о выходных или рабочих днях, и отразим среднее значение по времени заказов (выходные-будни)

```python
orders['Date'] = pd.to_datetime(orders['Date'])
orders['Weekends'] = orders['Date'].dt.weekday.apply(lambda x: 1 if x >= 5 else 0)
```

```python
# Вычисляем среднее значение Weekends для каждого ClientUUId
week_balance = orders.groupby('ClientUUId')['Weekends'].mean().reset_index(name='WeekBalance')
orders = orders.merge(week_balance, on='ClientUUId', how='left')
```

### Выведем наиболее распространённый ID пиццерии для каждого клиента

```python
unituuid_mode = orders.groupby('ClientUUId')['UnitUUId'].agg(lambda x: x.mode()[0]).reset_index(name='UnitUUIdMain')
orders = orders.merge(unituuid_mode, on='ClientUUId', how='left')
```

### Переносим данные о предпочтительной платформе

```python
platform_mode = mobile_events.groupby('ClientUUId')['Platform'].agg(lambda x: x.mode()[0]).reset_index(name='MostFrequentPlatform')
platform_mode['MostFrequentPlatform'] = platform_mode['MostFrequentPlatform'].map({'ios': 1, 'android': 0}).astype(int)
platform_mode.rename(columns={'MostFrequentPlatform': 'PlatformMain'}, inplace=True)
orders = orders.merge(platform_mode, on='ClientUUId', how='left')
```

### Сохраняем факт применения промоакции как среднее значение применения промо

```python
apply_promo = orders.groupby('ClientUUId')['apply_promo'].mean().reset_index(name='ApplyedPromoBalance')
orders = orders.merge(apply_promo, on='ClientUUId', how='left')
```
